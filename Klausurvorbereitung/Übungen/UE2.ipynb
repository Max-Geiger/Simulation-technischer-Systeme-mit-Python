```python
#Aufgabe 2.14

def someFunc(f1,f2,f3): 
    summe = f1+f2+f3
    produkt = f1*f2*f3 
    maximal = np.max([f1,f2,f3]) 
    minimal = np.min([f1,f2,f3])

    return summe,produkt,[maximal,minimal]


(sum, pro, [max,min]) = someFunc(2,3,4)
print(sum, pro, [max,min])
```
