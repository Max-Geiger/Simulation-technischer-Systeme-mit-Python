```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
from scipy.integrate import odeint
```

```python
#6.1
def func1(t,y):
    m=0.5
    d=0.25
    k=1

    y1p = y[1]
    y2p = -d/m*y[1]-k/m*y[0]
    return [y1p,y2p]

sol1 = solve_ivp(func1,t_span=(0,20),y0=[2,2],max_step=0.1)

plt.plot(sol1.t,sol1.y[0,:],'o-',label='solve_ivp')
plt.xlabel('t')
plt.ylabel('Auslenkung m')
plt.show()


```
