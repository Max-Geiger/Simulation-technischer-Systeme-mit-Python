```python
#A3.1
file_path = 'data_export.txt'
df = pd.read_csv(file_path, sep="-")
#df.to_dict()
#Zeit= df['Zeit'].tolist()
#df['Zeit'] = df.Zeit.apply(lambda x: x[1:-1].split(','))
# print(df)
# df.describe

def importData(Path): 
    infile = open(Path,'r') 
    lineStr = [] 
    for line in infile: 
        lineStr.append(line) 
    infile.close() 
    time_temp1 = lineStr[1].split(' ') # split label and numbers 
    data_temp1 = lineStr[2].split(' ') 
    time = [float(i) for i in time_temp1[1].split('-')] # convert number-string to list 
    data_temp2 = data_temp1[1].replace(',','.') 
    data = [float(i) for i in data_temp2.split('-')] 
    daten = {lineStr[0].strip(): {time_temp1[0]: time, data_temp1[0]: data}} # ohne .strip Operator wäre noch eine Leertaste hinter dem Wort 
    return daten 

daten = importData(file_path)
Zeit = daten['Probe1']['Zeit'] 
Wert = daten['Probe1']['Wert']

plt.plot(Zeit,Wert)
plt.grid()
plt.xlabel('Zeit') 
plt.ylabel('Wert')
plt.title('Probe 1')
```

```python
#A3.2
'''
Gegeben ist das zweite und fünfte Polynom der Taylorentwicklung einer Rechteckfunktion.
𝑦1(𝑡)=4𝜋sin(𝜔𝑡)+43𝜋sin(3𝜔𝑡) 
𝑦2(𝑡)=4𝜋sin(𝜔𝑡)+43𝜋sin(3𝜔𝑡)+45𝜋sin(5𝜔𝑡)+47𝜋sin(7𝜔𝑡)+49𝜋sin(9𝜔𝑡)
Plotten sie die Funktionen im Wertebereich [0,10] mit einer Auflösung von 0,001. Es gilt 𝜔=1
Zeigen Sie den Einfluss des Parameters 𝜔 in einem Fenster mithilfe von subplots.
'''
x = np.linspace(0, 10, 10000)
#print(10/0.001)
def funktion1(x,omega=1):
    return 4*np.pi*np.sin(x*omega) + 43*np.pi*np.sin(3*x*omega)

def funktion2(x,omega=1):
    return 4*np.pi*np.sin(x*omega) + 43*np.pi*np.sin(3*x*omega) + 45*np.pi*np.sin(5*x*omega) + 47*np.pi*np.sin(7*x*omega) + 49*np.pi*np.sin(9*x*omega)

fig, (ax1,ax2,ax3) = plt.subplots(3,1, figsize=(16,12))
ax1.plot(x, funktion1(x,1))
ax1.plot(x, funktion2(x,1))
ax1.grid()
ax1.set_xlabel('Time')
ax1.set_ylabel('Wert, omega=1')


ax2.plot(x, funktion1(x,0.5))
ax2.plot(x, funktion2(x,0.5))
ax2.grid()
ax2.set_xlabel('Time')
ax2.set_ylabel('Wert, omega=0,5')

ax3.plot(x, funktion1(x,2))
ax3.plot(x, funktion2(x,2))
ax3.grid()
ax3.set_xlabel('Time')
ax3.set_ylabel('Wert, omega=2')
```

```python
#A3.7


def V (x,D=0.0):
    return 1/np.sqrt((1-x)**2+(2*x*D)**2)

x = np.arange(0,5,0.001)

plt.figure(figsize=(12, 8))
plt.plot(x,V(x), color='r', label='D=0')
plt.plot(x,V(x,D=0.2), color='g', label='D=0.2')
plt.plot(x,V(x,D=0.4), color='y', label='D=0.4')
plt.plot(x,V(x,D=0.6), color='c', label='D=0.6')
plt.plot(x,V(x,D=0.8), color='m', label='D=0.8')
plt.plot(x,V(x,D=1), color='k', label='D=1')
plt.ylim(0.01, 10.0)
plt.title('Variation of D')
plt.ylabel('Vergrößerungsfunktion')
plt.xlabel('Frequnezverhältnis')

plt.legend()

```
